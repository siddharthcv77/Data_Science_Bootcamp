```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('boston_housing_data.csv')
```

```python
df.head()
```

#### 1. Create a line plot of ZN and INDUS in the housing data.

	a. For ZN, use a solid green line. For INDUS, use a blue dashed line.
	b. Change the figure size to a width of 12 and height of 8.
	c. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.

```python

# Change the style sheet
plt.style.use('ggplot')  

#Adjust the size 
plt.figure(figsize=(12, 8))
#For ZN, use a solid green line. For INDUS, use a blue dashed line
plt.plot(df['ZN'], color='green', label='ZN')
plt.plot(df['INDUS'], color='blue', linestyle='--', label='INDUS')
plt.legend()
plt.show()
```

#### 2. Create a bar chart using col1 and col2 of dummy data.

	a. Give the plot a large title of your choosing.
	b. Move the legend to the lower-left corner.
	c. Do the same thing but with horizontal bars.
	d. Move the legend to the upper-right corner.

```python
import matplotlib.pyplot as plt

# Vertical Bar Chart
ax = df.iloc[:, [0, 1]].plot(kind='bar', figsize=(12,8))

ax.set_title('CRIM and ZN values', fontsize=24)
ax.legend(loc='lower left')
plt.show()

# Horizontal Bar Chart
ax = df.iloc[:, [0, 1]].plot(kind='barh', figsize=(12,8))
ax.set_title('CRIM and ZN values (Horizontal)', fontsize=24)
ax.legend(loc='upper right')
plt.show()
```

#### 3. Create a histogram with pandas for using MEDV in the housing data.

	a. Set the bins to 20

```python
df['MEDV'].plot(kind='hist', bins=20, figsize=(12, 8))
plt.title("MEDV Histogram")
plt.xlabel("MEDV Value")
plt.ylabel("Frequency")
plt.show()
```

#### 4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
# Positive correlation: RM and MEDV
plt.scatter(df['RM'], df['MEDV'])
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.title('Scatter plot between RM and MEDV')
plt.show()

```

#### 5. Now, create a scatter plot of two heatmap entries that appear to have negative correlation.

```python
# Negative correlation: LSTAT and MEDV
plt.scatter(df['LSTAT'], df['MEDV'])
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('Scatter plot between LSTAT and MEDV')
plt.show()
```
